> Le code ci dessous donne un tensor PyTorch contenant les ids de tokens pour le texte donné. On peut également utiliser la méthode tokenize pour obtenir une liste de tokens plutôt qu'un tensor.⬇️

```
import pandas as pd

df_author_context = pd.read_csv('df_author_context.csv')

texts = df_author_context['context_left'].tolist()
```

```
from transformers import AutoTokenizer

# Choisir un modèle pré-entraîné
model_name = "facebook/bart-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenizez le texte
input_ids = [tokenizer.encode(text, padding ="max_length", truncation=True, max_length=50, return_tensors='pt') for text in texts]

print(input_ids)
```

> Le format de sortie de input_ids est une liste d'entiers (ou une tensor 1D dans le cas d'utilisation avec PyTorch). Cette liste représente les identifiants de chaque token (c'est-à-dire, chaque mot ou symbole pré-défini) dans le texte d'entrée, qui ont été préalablement mappé sur des entiers lors de l'entraînement du modèle de traitement du langage. Chaque entrée dans cette liste représente un token unique, et la position dans la liste indique l'ordre des tokens dans le texte d'entrée.
